```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data =pd.read_csv('cost_revenue_dirty.csv')
```

```python
data
```

```python
data.describe()
```

```python
# X =data['production_budget_usd']
# y =data['worldwide_gross_usd']
X =pd.DataFrame(data, columns=['production_budget_usd'])
y =pd.DataFrame(data, columns=['worldwide_gross_usd'])
```

# Plotting a scatter 

```python
plt.figure(figsize =(10,7))
plt.scatter(X, y, alpha =0.6)
plt.title('The budget vs the global revenue')
plt.ylabel('Worldwide Gross $')
plt.xlabel('Production Budget $')
plt.ylim(0,3000000000)
plt.xlim(0, 500000000)
plt.show()
```

```python
regression =LinearRegression()
```

# running our regression using the fit function

```python
regressor =regression.fit(X,y)
```

# finding the coeficient of the slop m

```python
regressor
```

```python
print('The coeficient of the slop is',regressor.coef_)
```

# the above coef means that for every 1.00 usd spent on a movie, there is a revenue of $ 3.111

```python
print('the intercept on the y-axis is', regressor.intercept_)
```

# For a budget of 0.00 usd, we loss by 7236192 usd

```python
plt.figure(figsize =(10,7))
plt.scatter(X, y, alpha =0.6)
plt.title('The budget vs the global revenue')
plt.ylabel('Worldwide Gross $')
plt.xlabel('Production Budget $')
plt.plot(X, regressor.predict(X), color='green', linewidth=6)
plt.ylim(0,3000000000)
plt.xlim(0, 500000000)
plt.show()
```

```python
rsquare =regression.score(X,y)
rsquare
```
